```python
%pylab inline
import json
from pprint import pprint
json_data = open("hnData/hn_saved_stories.json").read()
data = json.loads(json_data)
posts = [el['title'] for el in data['data']]
print("Total Posts: ",len(posts))
print("First 15...\n",posts[:15])
```

# Defining Models

## A. TF-IDF + Stemming

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk.stem
import scipy as sp

english_stemmer = nltk.stem.SnowballStemmer('english')

class StemmedTfidfVectorizer(TfidfVectorizer):
    def build_analyzer(self):
        analyzer = super(TfidfVectorizer, self).build_analyzer()
        return lambda doc: (english_stemmer.stem(w) for w in analyzer(doc))

vectorizer = StemmedTfidfVectorizer(min_df=1, stop_words='english', analyzer='word', ngram_range=(1,2))
```

##B. Dimensionality Reduction

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import Pipeline

lsa = TruncatedSVD(n_components=30)
normalizer = Normalizer(copy = True, norm='l2')
lsa_normalizer = Pipeline(steps = [
        ('lsa', lsa),
        ('normalizer', normalizer)
])
```

##C. K-Means

```python
from sklearn.cluster import KMeans
num_clusters = 25
km = KMeans(n_clusters=num_clusters, init='random', n_init=1)
```

#Fitting Models

#Predict

```python
new_post = "I like Machine Learning and I do it in python"
new_post_vec = vectorizer.transform([new_post])
new_post_d = lsa_normalizer.transform(new_post_vec)
new_post_label = km.predict(new_post_d)[0]

similar_indices = (km.labels_==new_post_label).nonzero()[0]
similar = []
for i in similar_indices:
    dist = sp.linalg.norm((new_post_d - X[i]))
    similar.append((dist, posts[i]))
similar = sorted(similar)
print "#Posts in cluster: ",len(similar)
print "Most similar in cluster : ", similar[0] # most similar in cluster
print "Mid similar in cluster  : ", similar[len(similar)/2] # mid-value in cluster
print "Least similar in cluster: ", similar[-1] # least similar in cluster
```

# Topic Modeling: LDA

After scraping all the links, model all of them using LDA


```python
# Generating multiple Kmeans clusters and running LDA to discover interesting ones.
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) #look at dependency issues later
from gensim import corpora, models, similarities
from nltk.corpus import stopwords

def process_corpora(text):
    tokens = nltk.word_tokenize(text)
    return [english_stemmer.stem(w) for w in tokens if w not in stopwords.words('english')]

texts = [process_corpora(p) for p in posts]
dic = corpora.Dictionary(texts)
corpus = [dic.doc2bow(text) for text in texts]
lda = models.LdaModel(corpus=corpus, id2word=dic, num_topics=100, passes=5, iterations = 10000,alpha=1.0)
```

```python
#lda.print_topics(50)
topics = [lda[c] for c in corpus]
plt.hist([len(t) for t in topics])
plt.ylabel('Nr of documents')
plt.xlabel('Nr of topics')
```
